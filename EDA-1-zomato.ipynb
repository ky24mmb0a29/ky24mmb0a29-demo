```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df = pd.read_csv("zomato.csv", encoding="latin1")
df
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cmap='viridis',cbar=True)
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
country_names=final_df.Country.value_counts().index
country_values=final_df.Country.value_counts().values
plt.pie(country_values[:3],labels=country_names[:3],
        autopct='%1.1f%%')
plt.show()

```

```python
df.columns
```

```python
ratings=df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,9)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings,hue='Rating color',palette=['grey','Red','Orange','yellow','Green','Green'])
plt.show()
```

```python
#COUNTRIES NAMES WHICH GAVE '0' RATINGS
names_Q1=final_df.groupby(['Aggregate rating','Country']).size()
names_Q1[0]
```

```python
#Corresponding country currencies
names_Q2=final_df.groupby(['Currency','Country']).size()
names_Q2
```

```python
#No.of Transactions only in Indian Cities
names_Q3=final_df.groupby(['City','Country']).size().reset_index().rename(columns={0:'CityCount'})
df_india=names_Q3[names_Q3['Country']=='India']
df_india=df_india.sort_values(by='CityCount',ascending=False)
df_City=df_india.City.tolist()
df_count=df_india.CityCount.tolist()
matplotlib.rcParams['figure.figsize']=(12,6)
plt.pie(df_count[:5],labels=df_City[:5],autopct='%1.1f%%')
```
