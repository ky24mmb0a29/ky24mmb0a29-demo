```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('supermarket_sales.xlsx')
df.head()
```

## Count plot

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Product line', data=df,hue='Gender',palette='autumn')
```

## Bar plot

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total',data=df,hue='Gender',palette='RdBu',ci=None,estimator='sum')
```

## Histplot

```python
sns.histplot(x='Total',data=df,bins=40,kde=True,element='poly',color='red',hue='Payment',multiple='stack')
plt.figure(figsize=(15,5))
```
